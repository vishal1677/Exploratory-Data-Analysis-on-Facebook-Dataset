```python
x=df[['age','dob_day','dob_year','dob_month','gender_male','gender_female','friend_count','friendships_initiated','likes','likes_received','mobile_likes','mobile_likes_received','www_likes','www_likes_received',]]
y=df['tenure']
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
from sklearn import metrics
print(metrics.mean_absolute_error(y_test,y_pred))
print(metrics.mean_squared_error(y_test,y_pred))
print(np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

```python
x1=df[['age', 'friend_count', 'friendships_initiated', 'likes', 'likes_received', 'gender']]
y1=df['tenure']
```

```python
from sklearn.model_selection import train_test_split

x1_train, x1_test, y1_train, y1_test = train_test_split(x1, y1, test_size=0.2, random_state=42)
```

```python
df=df.dropna()
df.isnull().sum()
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x1_train,y1_train)
y1_pred=lr.predict(x1_test)
```

```python
from sklearn import metrics
print(metrics.mean_absolute_error(y1_test,y1_pred))
print(metrics.mean_squared_error(y1_test,y1_pred))
print(np.sqrt(metrics.mean_squared_error(y1_test,y1_pred)))
```

```python
x2=df[['age', 'friend_count', 'friendships_initiated', 'likes', 'likes_received',]]
y2=df['gender']
```

```python
x2_train, x2_test, y2_train, y2_test = train_test_split(x2, y2, test_size=0.2, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier 
knn = KNeighborsClassifier(n_neighbors = 2)
knn.fit(x2_train, y2_train)
y2_pred = knn.predict(x2_test)
print(metrics.accuracy_score(y2_test,y2_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression

dtree = DecisionTreeClassifier(random_state=0)
model = LogisticRegression(random_state=0)
```

```python
dtree.fit(x2_train, y2_train)
y3_pred = dtree.predict(x2_test)

print("Accuracy = {}".format(metrics.accuracy_score(y_true=y2_test, y_pred=y3_pred)*100))
```
